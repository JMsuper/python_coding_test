```python
from itertools import combinations


def solution():
    global n, k, words, allSet
    
    if k < 5:
        return 0
    k -= 5
    alphas = set(['a','c','n','t','i'])

    remain_alphas = allSet - alphas
    if len(remain_alphas) <= k:
        return n

    # 비트 연산을 위한 딕셔너리 생성
    bit_dict = {}
    for alpha in remain_alphas:
        bit_dict[alpha] = len(bit_dict) 

    # 문자 set을 비트로 변환
    for i in range(len(words)):
        word = words[i] - alphas
        bit_word = 0b0
        for item in word:
            num = bit_dict[item]
            bit_word = bit_word | (1 << num)
        words[i] = bit_word


    # 가능한 모든 조합
    candidates = combinations(remain_alphas,k)
    
    ans = 0
    for candidate in candidates:
        cnt = 0
        candidate = set(candidate)
        bit_caditate = 0b0
        for item in candidate:
            num = bit_dict[item]
            bit_caditate = bit_caditate | (1 << num)

        for word in words:
            if word == bit_caditate & word :
                cnt += 1
        if cnt > ans:
            ans = cnt
    return ans


if __name__ == "__main__":
    n, k = map(int,input().split())
    words = []
    allSet = set({})
    for i in range(n):
        word = set(list(input()))
        words.append(word)
        allSet |= word

    print(solution())
```
