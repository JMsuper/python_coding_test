```python
r,c = map(int,input().split())

b_map = []
for i in range(r):
    temp = list(input())
    b_map.append(temp)

count = 0
for i in range(r):
    start = b_map[i][0]
    path = []
    if start == '.':
        x = 0
        y = i
        b_map[y][x]='O'
        path.append([y,x])
        while True:
            if x + 1 < c and y - 1 >= 0 and b_map[y - 1][x + 1] == '.':
                x += 1
                y -= 1
                b_map[y][x] = 'O'
                path.append([y,x])
                continue
            elif x + 1 < c and b_map[y][x + 1] == '.':
                x += 1
                b_map[y][x] = 'O'
                path.append([y,x])
                continue
            elif x + 1 < c and y + 1 < r and b_map[y + 1][x + 1] == '.':
                x += 1
                y += 1
                b_map[y][x] = 'O'
                path.append([y,x])
                continue
            else:
                if x != c-1:
                    for i in range(len(path)):
                        b_map[path[i][0]][path[i][1]] = '.'
                else:
                    count += 1
                break


print(count)
```

```python
# 애초에 문제를 잘못 접근하였다.
# dfs없이 그냥 할 수 있다고 생각했는데, dfs가 아니면 경로를 찾을 수 없는 게
# 존재했다
```

```python
r,c = map(int,input().split())

b_map = []
for i in range(r):
    temp = list(input())
    b_map.append(temp)

def dfs(y,x):
    stack = []
    stack.append([y,x])
    while stack:
        y, x = stack[-1]
        b_map[y][x] = 'O'
        if x == c - 1:
            return True
        if x + 1 < c and y - 1 >= 0 and b_map[y - 1][x + 1] == '.':
            x += 1
            y -= 1
            stack.append([y,x])
            continue
        elif x + 1 < c and b_map[y][x + 1] == '.':
            x += 1
            stack.append([y,x])
            continue
        elif x + 1 < c and y + 1 < r and b_map[y + 1][x + 1] == '.':
            x += 1
            y += 1
            stack.append([y,x])
            continue
        else:
            stack.pop()
    return False
        
answer = 0
for i in range(r):
    y = i
    x = 0
    if dfs(y,x):
        answer += 1

print(answer)
```
