```python
# 증말 어려웠던 문제
# 이진탐색, 우선순위큐를 사용하여 문제를 해결하려고 하였다.
# 우선순위큐에 게임판에서 탱크가 있어야 하는 타겟 행, 열을 집어넣어
# 해당 행, 열에 우선적으로 가장 가까이 있는 탱크를 움직이려고 하였다.
# 그러나, 어떠한 행, 열으로 이동해야 하는지에 대한 조건이 갈수록 복잡해졌다.
# 그리디 문제임에도 불구하고 너무 복잡한 접근으로 인해 실패하였다.

# 이 문제의 조건을 간단히 하면 다음과 같다.
# 한 행과 한 열에는 반드시 한 탱크만 있어야 한다.
# 한 공간에 탱크가 한 개만 존재해야 한다.
# 최소한의 움직임이라면 어떠한 경로로 움직이는 지는 상관없다.

# 나는 한 공간에 탱크가 한 개만 존재해야 하는 조건을 제대로 충족시키지 못하였다.
# 또한, 최소한의 움직임을 만들기 위한 기준을 제대로 설정하지 못하였다.
# 이 문제를 해결하기 위해서 N개의 탱크에 대해 N by N의 게임판이 있다는 것에 집중해야 했다.

# 참고링크 : https://blog.naver.com/pasdfq/221253664523
from bisect import bisect_left, bisect_right
import heapq

n = int(input())
loc_list = []
for i in range(n):
    loc = [i+1]
    loc.append(list(map(int,input().split(" "))))
    loc_list.append(loc)

print(loc_list)

n_map = [[0]*n for i in range(n)]

row_hq = []
column_hq = []
for i in range(n):
    row_hq.heappush([0,i+1])
    column_hq.heappush([0,i+1])

for i in range(n):
    name, row, column = loc_list[i]
    n_map[row-1][column-1] = name
    
    
```

```python
n = int(input())
loc_list = []
answer = []
for i in range(n):
    loc = []
    loc.append(list(map(int,input().split(" "))))
    loc[0][0] -= 1
    loc[0][1]-= 1
    loc.append(i+1)
    loc_list.append(loc)

loc_list.sort(key=lambda x:x[0][0])    
goUp = []
goDown = []

for i in range(n):
    if loc_list[i][0][0] > i:
        goUp.append(i)
    if loc_list[i][0][0] < i:
        goDown.append(i)

goDown.sort(key=lambda x:-x)
for i in goUp:
    x = i
    while x < loc_list[i][0][0]:
        answer.append([loc_list[i][1],'U'])
        x += 1
for i in goDown:
    x = loc_list[i][0][0]
    while x < i:
        answer.append([loc_list[i][1],'D'])
        x += 1

for i in range(n):
    loc_list[i][0][0], loc_list[i][0][1] = loc_list[i][0][1], loc_list[i][0][0]
    
loc_list.sort(key=lambda x:x[0][0])    
goLeft = []
goRight = []

for i in range(n):
    if loc_list[i][0][0] > i:
        goLeft.append(i)
    if loc_list[i][0][0] < i:
        goRight.append(i)

goRight.sort(key=lambda x:-x)
for i in goLeft:
    x = i
    while x < loc_list[i][0][0]:
        answer.append([loc_list[i][1],'L'])
        x += 1
for i in goRight:
    x = loc_list[i][0][0]
    while x < i:
        answer.append([loc_list[i][1],'R'])
        x += 1

print(len(answer))
for name, move in answer:
    print(name,move)
```
