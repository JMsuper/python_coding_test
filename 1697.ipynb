```python
# 걷기 : +1 or -1
# 순간이동 : *2
# +걷기, -걷기, *2걷기의 3가지 경우의 수가 존재한다.
# 너비우선탐색을 수행하며 k에 도달하는 때의 시간을 출력한다.

from collections import deque

n, k = map(int,input().split(' '))
answer = 0
queue = deque([[n,0]])

# 각각의 횟수를 카운팅 해야한다. how to?
def bfs():
    pos = queue.popleft()
#     print(pos)
#     print(queue)
    queue.append([pos[0]+1,pos[1]+1])
    queue.append([pos[0]-1,pos[1]+1])
    queue.append([pos[0]*2,pos[1]+1])
    return pos    

while True:
    pos = bfs()
    if pos[0] == k:
        answer = pos[1]
        break
print(answer)
```
