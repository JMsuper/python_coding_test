```python
#  : +1 or -1
#  : *2
# +, -, *2 3   .
#   k    .

from collections import deque

n, k = map(int,input().split(' '))
answer = 0
queue = deque([[n,0]])

#    . how to?
def bfs():
    pos = queue.popleft()
#     print(pos)
#     print(queue)
    queue.append([pos[0]+1,pos[1]+1])
    queue.append([pos[0]-1,pos[1]+1])
    queue.append([pos[0]*2,pos[1]+1])
    return pos    

while True:
    pos = bfs()
    if pos[0] == k:
        answer = pos[1]
        break
print(answer)
```
